## Missing Value Treatment  

```python
#Treating missing values in Train dataset
train_df['Applicant_City_PIN']=train_df['Applicant_City_PIN'].ffill().bfill()
train_df['Applicant_Gender']=train_df['Applicant_Gender'].ffill().bfill()
train_df['Applicant_BirthDate']=train_df['Applicant_BirthDate'].ffill().bfill()
train_df['Applicant_Marital_Status']=train_df['Applicant_Marital_Status'].ffill().bfill()
train_df['Applicant_Occupation']=train_df['Applicant_Occupation'].ffill().bfill()
train_df['Applicant_Qualification']=train_df['Applicant_Qualification'].ffill().bfill()
train_df['Manager_DOJ']=train_df['Manager_DOJ'].ffill().bfill()
train_df['Manager_Joining_Designation']=train_df['Manager_Joining_Designation'].ffill().bfill()
train_df['Manager_Current_Designation']=train_df['Manager_Current_Designation'].ffill().bfill()
train_df['Manager_Grade']=train_df['Manager_Grade'].ffill().bfill()
train_df['Manager_Status']=train_df['Manager_Status'].ffill().bfill()
train_df['Manager_Gender']=train_df['Manager_Gender'].ffill().bfill()
train_df['Manager_DoB']=train_df['Manager_DoB'].ffill().bfill()
train_df['Manager_Num_Application']=train_df['Manager_Num_Application'].ffill().bfill()
train_df['Manager_Business']=train_df['Manager_Business'].fillna(np.mean(train_df['Manager_Business']))
train_df['Manager_Num_Products']=train_df['Manager_Num_Products'].ffill().bfill()
train_df['Manager_Business2']=train_df['Manager_Business2'].fillna(np.mean(train_df['Manager_Business2']))
train_df['Manager_Num_Products2']=train_df['Manager_Num_Products2'].ffill().bfill()
train_df['Manager_Num_Coded']=train_df['Manager_Num_Coded'].ffill().bfill()
```

```python
# Treating missing values in Test dataset
test_df['Applicant_City_PIN']=test_df['Applicant_City_PIN'].ffill().bfill()
test_df['Applicant_Gender']=test_df['Applicant_Gender'].ffill().bfill()
test_df['Applicant_BirthDate']=test_df['Applicant_BirthDate'].ffill().bfill()
test_df['Applicant_Marital_Status']=test_df['Applicant_Marital_Status'].ffill().bfill()
test_df['Applicant_Occupation']=test_df['Applicant_Occupation'].ffill().bfill()
test_df['Applicant_Qualification']=test_df['Applicant_Qualification'].ffill().bfill()
test_df['Manager_DOJ']=test_df['Manager_DOJ'].ffill().bfill()
test_df['Manager_Joining_Designation']=test_df['Manager_Joining_Designation'].ffill().bfill()
test_df['Manager_Current_Designation']=test_df['Manager_Current_Designation'].ffill().bfill()
test_df['Manager_Grade']=test_df['Manager_Grade'].ffill().bfill()
test_df['Manager_Status']=test_df['Manager_Status'].ffill().bfill()
test_df['Manager_Gender']=test_df['Manager_Gender'].ffill().bfill()
test_df['Manager_DoB']=test_df['Manager_DoB'].ffill().bfill()
test_df['Manager_Num_Application']=test_df['Manager_Num_Application'].ffill().bfill()
test_df['Manager_Business']=test_df['Manager_Business'].fillna(np.mean(test_df['Manager_Business']))
test_df['Manager_Num_Products']=test_df['Manager_Num_Products'].ffill().bfill()
test_df['Manager_Business2']=test_df['Manager_Business2'].fillna(np.mean(test_df['Manager_Business2']))
test_df['Manager_Num_Products2']=test_df['Manager_Num_Products2'].ffill().bfill()
test_df['Manager_Num_Coded']=test_df['Manager_Num_Coded'].ffill().bfill()
```
